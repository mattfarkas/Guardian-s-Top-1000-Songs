```python
import pandas as pd
import numpy as np
```

```python
main_data = pd.read_csv("top_songs_more_info.csv")
genre_data = pd.read_csv("artists_and_genres.csv")
unemployment = pd.read_csv("unemployment.csv")
```

```python
main_data.drop(['Unnamed: 0', 'Unnamed: 0.1'], inplace=True, axis=1)
genre_data.drop(['Unnamed: 0', 'Unnamed: 0.1'], inplace=True, axis=1)
```

```python
years = main_data['YEAR'].unique().tolist()
years.sort()
genres = genre_data['master_genre'].unique().tolist()
genres.sort()
_ = genres.pop(0)
```

```python
combined = main_data.merge(genre_data, how='inner', left_on='ARTIST', right_on='artist')
```

```python
genre_pop_by_year = pd.DataFrame()
for year in years:
    for genre in genres:
        enum_dic = {}
        subset = combined[(combined['YEAR'] == year) & (combined['master_genre'] == genre)]
        enum_dic['Genre'] = genre
        enum_dic['Year'] = year
        total_gen_year = float(len(combined[(combined['YEAR'] == year) & (combined['master_genre'].notnull())]))
        enum_dic['Popularity'] = len(subset) / total_gen_year
        enum_dic['Spotify_Popularity'] = subset['song_pop'].mean()
        enum_dic['Acousticness'] = subset['acousticness'].mean()
        enum_dic['Energy'] = subset['energy'].mean()
        enum_dic['Instrumentalness'] = subset['instrumentalness'].mean()
        enum_dic['Loudness'] = subset['loudness'].mean()
        enum_dic['Tempo'] = subset['tempo'].mean()
        enum_dic['Valence'] = subset['valence'].mean()
        enum_dic['Danceability'] = subset['danceability'].mean()
        
        genre_pop_by_year = genre_pop_by_year.append(enum_dic, ignore_index=True)
genre_pop_by_year['Year'] = genre_pop_by_year['Year'].astype(int)
genre_pop_by_year
```

```python
with_unemployment = genre_pop_by_year.merge(unemployment, how='left', on='Year')
```

```python
with_unemployment
```

```python
with_unemployment.to_csv("genre_by_year_breakdown.csv")
```
